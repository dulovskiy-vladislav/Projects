```python
import pandas as pd
```

```python
data = pd.read_csv('product_gallery.csv')
```

```python
data.drop(columns='CategoryID', inplace = True)
data.drop(columns='EmployeeID', inplace = True)
data.drop(columns='Fax', inplace = True)
data.drop(columns='ProductID', inplace = True)
data.drop(columns='ShipperID', inplace = True)
data.drop(columns='SupplierID', inplace = True)
```

```python
#Переделаем название столбцов под условие Naming convention
data.columns = ['address', 'category', 'city', 'customer_company','shippers_company', 'customer', 'country', \
                'order_date', 'order_id', 'phone', 'postal_code', 'product', 'state_province', 'discount', 'freight', \
                'quantity', 'qunatity_per_unit', 'unit_cost', 'unit_price', 'unit_price_products', 'units_in_stock', \
                'unit_on_order', 'sales', 'profit', 'city_and_country']
```

```python
data['order_date']= pd.to_datetime(data['order_date'], format='%d.%m.%Y')
```

```python
data['discount'] = data['discount'].str.replace(',', '.').astype(float)
data['sales'] = data['sales'].str.replace(',', '.').astype(float)
data['freight'] = data['freight'].str.replace(',', '.').astype(float)
data['profit'] = data['profit'].str.replace(',', '.').astype(float)
data['unit_cost'] = data['unit_cost'].str.replace(',', '.').astype(float)
data['unit_price'] = data['unit_price'].str.replace(',', '.').astype(float)
data['unit_price_products'] = data['unit_price_products'].str.replace(',', '.').astype(float)
```

```python
import pandas_profiling
pandas_profiling.ProfileReport(data)
```

```python
data['state_province'].value_counts()
```

```python
#Расшифруем код каждого штата
states = {
    'AK': 'Alaska',
    'BC': 'British Columbia',
    'CA': 'California',
    'IL': 'Illinois',
    'ID': 'Idaho',
    'MT': 'Montana',
    'NM': 'New Mexico',
    'OR': 'Oregon',
    'QC': 'Quebec',
    'WA': 'Washington',
    'WY': 'Wyoming'
}
data['state_province'] = data['state_province'].map(states)
```

```python
data['state_province'].value_counts()
```

```python
#Создадим базу данных из получившегося DataFrame
import sqlite3
conn = sqlite3.connect('DataBase Product Gallery')
data.to_sql('product_gallery', conn, if_exists='replace', index = False)
conn.close()
```
