```python
import pandas as pd
```

```python
data = pd.read_csv('movie_dataset.csv', index_col= ['index'])
```

```python
data
```

```python
data.drop(columns = 'crew', inplace = True)
data.drop(columns = 'production_companies', inplace = True)
data.drop(columns = 'spoken_languages', inplace = True)
data.drop(columns = 'production_countries', inplace = True)
```

```python
data.dropna(subset=['genres'],axis = 0, inplace=True)
```

```python
data['release_date']= pd.to_datetime(data['release_date'])
```

```python
data['genres'] = data['genres'].apply(lambda x: x.split(' ')[0])
```

```python
data = data[data['status'] == 'Released']
```

```python
from iso639 import languages
```

```python
def get_language_name(iso_code):
    try:
        language_name = languages.get(alpha2=iso_code).name
    except KeyError:
        language_name = "Unknown"
    
    return language_name

data['original_language'] = data['original_language'].apply(lambda x: get_language_name(x))
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('project_datastudy.sqlite')
data.to_sql('movie_base', conn, if_exists='replace', index = False)
conn.close()
```
